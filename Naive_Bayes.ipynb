```python
# PREPROCESSING DATA

from sklearn.preprocessing import LabelEncoder
for col in ['gender','ever_married','work_type','Residence_type','smoking_status'] :
    label_encoder = LabelEncoder()
    tampil[col] = label_encoder.fit_transform(tampil[col])
```

```python
tampil
```

```python
# Memisahkan kolom targer (stroke) dari dataset
x = tampil.drop ('stroke', axis=1)
y = tampil ['stroke']
```

```python
x
```

```python
y
```

```python
# Splitting data tranning dan testing
from sklearn.model_selection import train_test_split
```

```python
# 85:15
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.15)
```

```python
x_train.shape
```

```python
## MODELING
# Naive Bayes

from sklearn.naive_bayes import GaussianNB
modelNB = GaussianNB ()
modelNB.fit (x_train, y_train)
```

```python
# Testing model untuk melakukan Prediksi
hasil_prediksi = modelNB.predict (x_test)
```

```python
# Cek keakuratan hasil prediksi
from sklearn.metrics import accuracy_score
print ('Akurasi Model   : ', accuracy_score (y_test, hasil_prediksi))
```

```python
pasien_baru = ([[1,20,0,0,0,3,1,150,69,2]])
hasil_prediksi_pasien = modelNB.predict (pasien_baru)
print (f"Hasilnya Adalah ", hasil_prediksi_pasien)
```

```python
#### Desision Tree
from sklearn.tree import DecisionTreeClassifier
modelDT = DecisionTreeClassifier (max_depth = 5)
modelDT.fit (x_train, y_train)
```

```python
hasil_prediksi_dt = modelDT.predict (x_test)
print ('Akurasi Model : ', accuracy_score (y_test, hasil_prediksi_dt)*100)
```

```python
# Visualisasi model Decision Tree
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
plt.subplots (figsize=(40,20))
plot_tree(modelDT, fontsize = 20)
plt.show()
```
